# Loading and Displaying Images

```python
# We need this function to open tif-files.
from tifffile import imread

# we will use this to display the image
from matplotlib import pyplot as plt

# we will use this to access and manipulate the image
import numpy as np
```

```python
# opening the file
galaxy= imread('/mnt/terajug/m51.tif')
```

```python
plt.imshow(galaxy)
```

```python
# Let us have a look the dimensions of the image
print(galaxy.shape)
# Note that y is listed before x
```

```python
# Let us have a look at the pixel values
print(np.min(galaxy), np.max(galaxy), np.mean(galaxy), np.std(galaxy))
```

```python
# We want to adjust the brightness and contrast used for displaying.
# Using percentiles is usually a good idea to find asuitable values 
# for the lowest and highest values to be displayed.
low=np.percentile(galaxy,5)
high=np.percentile(galaxy,99)
print(low,high)
# vmin and vmax set the lowest and highest values
plt.imshow(galaxy,vmin=low, vmax=high)
```

```python
# we can use a differenct color map to display the image
low=np.percentile(galaxy,5)
high=np.percentile(galaxy,99)
plt.imshow(galaxy,vmin=low, vmax=high, cmap='magma')
plt.show()

low=np.percentile(galaxy,5)
high=np.percentile(galaxy,99)
plt.imshow(galaxy,vmin=low, vmax=high, cmap='gray')
plt.show()
```

# Manipulating Images and Pixels

```python
# we can apply arithmetic operations such as inverting the image 
galaxy_=galaxy * -1

low=np.percentile(galaxy_,5)
high=np.percentile(galaxy_,99.99)
print(low,high)
plt.imshow(galaxy_,vmin=low, vmax=high)
```

```python
# we can apply arithmetic operations such as a gamma correction 
# (the '**' stands for exponentiation)
galaxy_=galaxy ** 0.1

low=np.percentile(galaxy_,5)
high=np.percentile(galaxy_,99.99)
print(low,high)
plt.imshow(galaxy_,vmin=low, vmax=high)
```

```python
plt.imshow(galaxy)
plt.show()

# we can access a region of interest
# we first have to provide y, then x-coordinates
# 'a:b' gives us everything between 'a' and 'b'
# ':a' everything until 'a'
# 'a:' you can figure out yourself
# you can also use negative indexes. Try it.
galaxy_=galaxy[200:300,100:200]
plt.imshow(galaxy_)
plt.show()
```

```python
# we can manipulate individual pixels 
galaxy_[20,20]= 10000
plt.imshow(galaxy_)
plt.show()

# Note that this also effects the original image
plt.imshow(galaxy[200:300,100:200])
plt.show()
```

```python
# if we do not want this effect, we can make a copy 
galaxy_copy= galaxy_.copy()
galaxy_copy[30,30]= 10000
plt.imshow(galaxy_copy)
plt.show()

# Manipulating the copy does not affect the original
plt.imshow(galaxy[200:300,100:200])
plt.show()
```

```python
# We can also manipulate subimages
galaxy_copy= galaxy_.copy()
galaxy_copy[20:50,20:50]= 10000
plt.imshow(galaxy_copy)
plt.show()
```
