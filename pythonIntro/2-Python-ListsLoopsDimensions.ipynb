# Lists and Loops

```python
# We need this function to open tif-files.
from tifffile import imread

# We need this function to write tif-files.
from tifffile import imwrite

# we will use this to display the image
from matplotlib import pyplot as plt

# we will use this to access and manipulate the image
import numpy as np
```

```python
from glob import glob
# glob produces a list of files-paths fitting the expression
files= glob('/mnt/terajug/data/pythonIntro/head/*.tif')
```

```python
# to look at the paths we use a for loop over all files
# note that the files are not sorted
for f in files:
    print(f)
    
```

```python
# Let's sort the files
for f in sorted(files):
    print(f)
```

```python
# what if we only want the first n
for i in range(10):
    print(i, files[i])
    # let's load the files an display them
    plt.imshow(imread(files[i]))
    plt.show()
```

# Working with Lists and NP Arrays

```python
# we create an empty list
images=[]

# we will load all images an append them to the list
for f in sorted(files):
    print(f)
    images.append(imread(f))
    
# Let's see how many entries our list has
print(len(images))
```

```python
# We can make a numpy array from the list
head=np.array(images)

# Let's look at its dimensionality
# the dimensions are z, y, x
print(head.shape)

# saving our 3D tiff stack
# we also convert it 32 bit
# this is sometimes useful because fiji cannot open 64 bit tifs
imwrite('headStack.tif', head.astype(np.float32))
```

```python
# we show the maximum projection along the first axis.
plt.imshow(np.max(head,axis=0))

# we can also look at the average projection
#plt.imshow(np.mean(head,axis=0))

# we draw a horizontal line on top of the image
# we will make a cut at this position
ycut=122
plt.plot([0,256],[ycut,ycut],color='white')


# we draw a vertical line on top of the image
# we will make a cut at this position
xcut=125
plt.plot([xcut,xcut],[0,256],color='white')

# we are setting the range manually.
# if we don't, the result will look a little ugly
plt.xlim((0,256))
plt.ylim((256,0))
plt.show()
```

```python
# Let us look at one zx-slice
zxSlice=head[:,ycut,:]
print(zxSlice.shape)

plt.imshow(zxSlice)
plt.show()
```

```python
# Let us look at one xz-slice
zySlice=head[:,:,xcut]

plt.imshow(zySlice)
```

```python
# take a look at the swap axis function
# it will help us to make a yz-Slice
? np.swapaxes
```

```python
yzSlice=np.swapaxes(head,0,1)[:,:,xcut]
plt.imshow(yzSlice)
```

# More Loops

```python
# Here we draw a circle using for loops and if statements
centerX= 50
centerY= 50
for y in range(zxSlice.shape[0]):
    for x in range(zxSlice.shape[0]):
        if np.sqrt ( (x-centerX)**2 + (y-centerY)**2 ) < 20 :
            zxSlice[y,x]=1000
plt.imshow(zxSlice)
```
